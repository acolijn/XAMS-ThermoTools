# Heat Transfer Estimate into a 1/4″ Steel Tube Filled with Liquid Xenon

We want the heat leak into a 1/4″ stainless steel tube (length 0.8 m, filled with LXe at –100 °C)  
when it is surrounded by xenon gas at 0 °C and ~1 bar.

---

## Geometry and conditions

- Outer diameter: $D_o = 6.35\ \text{mm}$
- Wall thickness: ~0.9 mm → Inner diameter $\approx 4.6\ \text{mm}$
- Length: $L = 0.8\ \text{m}$
- Inside: LXe at $T_{\text{in}} = 173\ \text{K}$ (–100 °C)
- Outside: Xe gas at $T_{\text{out}} = 273\ \text{K}$ (0 °C)
- Temperature difference: $\Delta T \approx 100\ \text{K}$

Outer surface area:
$$
A_o = \pi D_o L \approx 0.0160\ \text{m}^2
$$

---

## Outside convection (dominant)

For natural convection around a horizontal cylinder, Churchill–Chu correlation:

$$
\text{Nu} = \left[0.60 + \frac{0.387\,\text{Ra}^{1/6}}
{\left(1 + (0.559/\text{Pr})^{9/16}\right)^{8/27}}\right]^2
$$

with
$$
h_o = \frac{\text{Nu}\,k}{D_o}
$$

Xenon gas properties near 273 K, 1 bar (approx):
- $k \approx 5.6\times10^{-3}\ \text{W m}^{-1}\text{K}^{-1}$
- $\mu \approx 2.3\times10^{-5}\ \text{Pa . s}$
- $\rho \approx 5.5\ \text{kg m}^{-3}$
- $\text{Pr} \approx 0.66$

These give $\text{Ra}\sim 3.5\times10^4$ and
$$
h_o \approx 5.2\ \text{W m}^{-2}\text{K}^{-1}.
$$

So the convective heat transfer is
$$
Q_{\text{conv}} = h_o A_o \Delta T
\approx 5.2 \times 0.0160 \times 100
\;\approx\; 8.3\ \text{W}.
$$

---

## Wall conduction

Cylindrical conduction resistance:
$$
R_{\text{wall}} = \frac{\ln(r_o/r_i)}{2\pi k_{\text{wall}} L}
\approx 0.0043\ \text{K/W}
$$

for stainless steel with $k_{\text{wall}}\approx 15\ \text{W m}^{-1}\text{K}^{-1}$.  
This is negligible compared with the outside film resistance
($1/(h_o A_o) \approx 12.5\ \text{K/W}$).

---

## Inside LXe convection

Even if the LXe side coefficient is modest, say $h_i=100$–$500\ \text{W m}^{-2}\text{K}^{-1}$,  
the resistance
$$
R_i = \frac{1}{h_i A_i}
$$

with $A_i\approx 0.0116\ \text{m}^2$ is between 0.86 and 0.17 K/W,  
still small compared with the outside film resistance. So the inside resistance is also negligible.

---

## Radiation

Linearized radiation coefficient at mean surface temperature $T_m\approx 248\ \text{K}$:

$$
h_{\text{rad}} = 4 \varepsilon \sigma T_m^3
$$

- Polished steel, $\varepsilon \approx 0.2$: $h_{\text{rad}}\approx 0.7\ \text{W m}^{-2}\text{K}^{-1}$  
  → $Q_{\text{rad}} \approx 1.1\ \text{W}$
- Dull/oxidized steel, $\varepsilon \approx 0.6$: $h_{\text{rad}}\approx 2.1$  
  → $Q_{\text{rad}} \approx 3.2\ \text{W}$

---

## Result

- Convection only: $\;\approx 8\ \text{W}$
- Convection + radiation (polished): $\;\approx 9\ \text{W}$
- Convection + radiation (oxidized): $\;\approx 11\ \text{W}$

---

**Bottom line:** About **8–11 W** of heat will leak into the liquid xenon through this tube,
depending on the surface emissivity.
